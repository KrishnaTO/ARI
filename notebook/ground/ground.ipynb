```python
def create_link(curie):

    # Examples    
    # https://www.ebi.ac.uk/ols4/ontologies/doid/classes/http%253A%252F%252Fpurl.obolibrary.org%252Fobo%252FDOID_0110751
    # https://www.ebi.ac.uk/ols4/ontologies/doid/classes/http%253A%252F%252Fpurl.obolibrary.org%252Fobo%252FDOID_13378
    # https://www.ebi.ac.uk/ols4/ontologies/uberon/classes/http%253A%252F%252Fpurl.obolibrary.org%252Fobo%252FUBERON_0000178
    
    curie_split = curie.split(':')
    ns = curie_split[0]
    id = curie_split[1]
    return f"https://www.ebi.ac.uk/ols/ontologies/{ns.lower()}/classes/http%253A%252F%252Fpurl.obolibrary.org%252Fobo%252F{ns}_{id}"
    
def ground_column(df, column, grounder, target_type="curie"):
    grounder.ground_df(df, column, target_column=f"{column}_reference", target_type="reference")

    if "curie" in target_type:
        grounder.ground_df(df, column, target_column=f"{column}_curie", target_type="curie")
        
        df[f"{column}_curie"] = df[f"{column}_curie"].apply(
            lambda x: x if pd.isna(x) else (x.upper() if ':' not in x else x.split(':')[0].upper() + ':' + x.split(':')[1])
        )
        
        # Add link
        df[f"{column}_link"] = df[f"{column}_curie"].apply(
            lambda x: create_link(x) if x and ':' in x else None
        ) 
        
        df[f"{column}_curie"] = df[f"{column}_curie"].apply(
            lambda x: f"{x.split(':')[0]}:{x.split(':')[1].zfill(7)}" if x and ':' in x else x
        )

        
    if "name" in target_type:
        df[f"{column}_name"] = df[f"{column}_reference"].apply(lambda x: x.name if x else None)
    return df[[ column ] + [ f"{column}_{t}" for t in target_type ] + [ f"{column}_link" ] if "curie" in target_type else [] ]
```
